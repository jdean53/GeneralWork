```python
import pandas as pd
import numpy as np
import FinOpsCodeDeck as finops
```

```python
sheet1 = pd.read_excel('Final_Annual_Returns_BBB.xlsx',sheet_name='Annual Prices',index_col='Year')
sheet2 = pd.read_excel('Final_Annual_Returns_BBB.xlsx',sheet_name='Annual Returns',index_col='Year').iloc[:,:3]
```

```python
'''
Problem 4 part A
'''
sheet1.pct_change().dropna()
```

```python
'''
problem 4 part B
'''
years = sheet1.index.max()
div_per_year = 100 / years
adj_px_asset_one = [sheet1['Asset 1'].loc[i] + (i)*div_per_year for i in range(years+1)]
sheet1['Asset 1'] = adj_px_asset_one
sheet1.pct_change().dropna()
```

```python
'''
problem 4 part c
'''
print('Expected Returns')
print(sheet2.mean())
print(' ')
print('Covariance Matrix')
print(sheet2.cov())
```

```python
'''
Problem 4 part d
'''
w = np.array([1/3,1/3,1/3])
rets = (w * sheet2).transpose().sum().to_numpy()
exp_ret = sum(w * sheet2.mean())
port_var = sum(sum(w * sheet2.cov().to_numpy()) * w)
port_sv = finops.downside_semi_variance(rets)
VaR = rets[np.percentile(rets,q=(0.1250)*100) < rets].min()
print('Part i: The expected return is {:.2f}%'.format(exp_ret))
print('Part ii: The portfolio variance is {:.2f}'.format(port_var))
print('Part iii: The downside semivariance is {:.2f}'.format(port_sv))
print('part iv: the VaR of the portfolio is {:.2f}'.format(VaR))
```
